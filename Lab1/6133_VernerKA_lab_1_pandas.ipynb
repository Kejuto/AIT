Для заполнения пропусков средними значениями, нужно вычислить среднее значение. И на их основе заполнить пропуски.

```python
# Для вычисление средних значений и замена ими средних значений df.mean()
# Для вставки этих значений нужно параметр полученный с df.mean() вставить в df.fillna
# В моем датасете первая колонка с датами имеет формат str для высчитывания mean в датасете и для этой строки нужно перевести в формат date
df['Date'] = pandas.to_datetime(df['Date'], format='%Y-%m-%d')
mean = df.mean()
df.fillna(mean, inplace=True)
df.head(15)
```

Значения успешно замененны средним по колонке

Далее нужно создать новое поле вычисленное на основе значений дргих полей

```python
# Создадим новую колонку являющиеся разницей между максимальным и минимальным значением

df['difference'] = df['High'] - df['Low']
df.head()
```

```python
# Через DataFramae.apply

df['total'] = df.apply(lambda row: row['Volume'] / row['Close'], axis=1)
df.head()
```

```python
# Через Series.apply

df['max_total'] = df['Volume'].apply(lambda x: x / df.loc[df['Volume'] == x, 'Low'].iloc[0])
df.head()
```

Решейпинг данныхх 1D --> 2D с использованием .pivot
Перед этим предварительно сгрупиируем данные по двум колонкам

```python
# Предварительно группируем данные
groping = df.groupby(['Date', 'Open'])['Close'].sum().reset_index()
print(groping)
```

```python
# Далее преобразовываем данные с помощью pivot()
piv_table = groping.pivot(index='Date', columns='Open', values='Close')
print(piv_table)
```

Так как для каждой даты уникальное значение Open, то вывод получается такой

```python
pivot_table = df.pivot_table(
    index = ['Date', 'Volume'],
    values = ['Open','Close', 'High'],
    aggfunc = 'count'
 )
print(pivot_table)
```

Метод .iterrows() в pandas позволяет проитерироваться по строкам DataFrame построчно. Каждая строка возвращается как пара (index, Series)

```python
# Итерируем по строкам в цикле for
for index, row in df.iterrows():
    # Вычисляем новый столбец как Low + High нвзовем ее Summing
    summing = row['Low'] + row['High']
    print(f"Index: {index}, Date: {row['Date']}. Summing: {summing}" )
```

MultiIndex - это структура данных в pandas, которая позволяет создавать иерархические индексы (несколько уровней группировки) для строк или столбцов DataFrame.

```python
# Цены по датам на акции, максимальные, минимальные
# Переставляем уровни
swapped_df = pivot_table.swaplevel(i=0, j=1)  # меняем уровни 0 и 1
print("\nПосле перестановки уровней:")
print(swapped_df)
```

```python
# Транспонируем матрицу

transpored = swapped_df.T.unstack()
print(transpored)
```

```python
# Добавление нового уровня
df_delite = pivot_table.reset_index('Volume', drop = True)
print("\n DataFrame после удаления уровня:")
print(df_delite)
```

Для демонстрации merge, разделим датасет на 2 части

```python
# Разделяеи от середнее значение в Date
mean_date = df['Date'].mean()
print(mean_date)
```

```python
part1 = df[df['Date'] <= mean_date]
part2 = df[df['Date'] > mean_date]

# Объединением по Open
merged = pandas.merge(part1, part2, on='Open', how='inner')
print(merged)
```

```python
# Используем горизонтальное объединение с помощью concat()

vertical_concat = pandas.concat([part1, part2], axis=0, ignore_index = True)

print(vertical_concat)
```
